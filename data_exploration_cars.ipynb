```python

# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.columns
```

```python
df.tail(5)
```

```python
df.dtypes
```

```python
df1 =  df.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Size'],axis =1 )
```

```python
df1.columns
```

```python
df1.head()
```

```python
# renaming the columns
df1 = df1.rename(columns ={'Engine HP':'HP','Engine Cylinders':'Cylinders','Transmission Type':'Transmission',
                           'Driven_Wheels':'Driven Mode','highway MPG':'MPG-H','city mpg': 'MPG-C', 'MSRP': 'Price' })
```

```python
df1.head()
```

```python
# dropping the duplicate rows
print('size before removing the duplicate data:',df1.shape)
```

```python
duplicate_rows = df1[df1.duplicated()]
print('size of  the duplicate data:',duplicate_rows.shape)
```

```python
df1.count()
```

```python
df2 = df1.drop_duplicates()
df2.shape
```

```python
df2.count()
```

```python
# finding the null values
df2.isnull().sum()
```

```python
df3 = df2.dropna()
```

```python
df3.shape
```

```python
df3.count()
```

```python
# checking if any null value in the data
df3.isnull().sum()
```

```python
cols = df3.columns.values
numeric_cols = df3.select_dtypes('number').columns
numeric_cols

```

```python
# finding the outliers
for i in numeric_cols:
  plt.subplot(1,2,1)
  sns.boxplot(x =df3[i])
  plt.show()
```

```python
# finding outliers using IQR
Q1 = df3[numeric_cols].quantile(0.25)
Q3 = df3[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# removing the outliers
df3.shape
```

```python
df4 = df3[~((df3 < (Q1 - 1.5*IQR)) | (df3 > (Q3 + 1.5*IQR))).any(axis = 1)]
df4.shape
```

```python
df4.columns
```

```python
# plotting histograms
cols
```

```python
for i in cols:
  plt.subplot(1,2,1)
  df4[i].value_counts().nlargest(50).plot(kind = 'bar',figsize=(20,5))
  plt.ylabel('Number')
  plt.xlabel(i)
  plt.show()
```
